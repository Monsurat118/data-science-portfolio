# Detect Outliers Numerically Using IQR
 What are we doing?
We will quantify outliers using the Interquartile Range (IQR) method.

```python
# Function to detect outliers using IQR
def detect_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)  # First quartile
    Q3 = df[column].quantile(0.75)  # Third quartile
    IQR = Q3 - Q1  # Interquartile range
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    return outliers

# Check the number of outliers per column
for col in num_cols:
    outliers = detect_outliers_iqr(df_filtered, col)
    print(f"{col}: {len(outliers)} outliers detected")
```

Displays the number of outliers per column.
Helps decide whether to remove or adjust extreme values.

# Remove Outliers Using IQR
What are we doing?
Now that we have identified outliers, we will remove them.

```python
# Function to remove outliers using IQR
def remove_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)  # First quartile
    Q3 = df[column].quantile(0.75)  # Third quartile
    IQR = Q3 - Q1  # Interquartile range
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

# Apply outlier removal for each numerical column
for col in num_cols:
    df_filtered = remove_outliers_iqr(df_filtered, col)

# Check summary statistics after outlier removal
df_filtered.describe()
```

Extreme values removed, making data more reliable.
Distributions become more normal, improving ML predictions.
